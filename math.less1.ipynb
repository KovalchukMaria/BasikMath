```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 101)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 101)
y = np.cos(x)**2 - np.sin(x)**2
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = 4 * np.linspace(-1, 1, 101)
y = np.sin(x)
plt.plot(x, y)
plt.grid()
plt.show()
```
